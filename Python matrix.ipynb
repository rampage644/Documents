```python
from IPython.display import HTML, display_html
import html
import math
```

```python
def table(list, columns=1):
    h = html.HTML()
    t = h.table
    rows = int(math.ceil(1.0*len(list)/columns))
    for i in range(rows):
        r = t.tr
        for j in range(columns):
            if len(list) > rows*j+i:
                r.td(list[rows*j+i][0], id=list[rows*j+i][1])
    return str(h)
```

```python
l1 = [(w, 'w') for w in work]
l2 = [(e, 'e') for e in exp]
l3 = [(u, 'u') for u in uns]
```

```python
display_html(table(l1+l2+l3, columns=7), raw=True)
```

## Unsupported modules

There is several reasons for some standard modules are unsupported:
* Single-threaded sandbox execution
* No intended GUI
* Sandbox session is one process which has no ability to spawn/fork
* No explicit network access
* Not compiled-in modules
* Posix-oriented process (not Windows/MacOS/Sun/IRIX)
* Determenism caused ZeroVM/Zrt constraints

Unsupported modules distribution is following:

```python
reasons = ['Multithreading', 'GUI', 'Multiprocessing', 'Networking', 'No modules', 'Windows/MacOS', 'Others']
modules = [2, 9, 7, 21, 5, 33, 18]
```

```python
fig, ax = plt.subplots(figsize=(6, 6))  # make the plot square
pie = ax.pie(modules, labels=reasons)
```
